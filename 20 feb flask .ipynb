Q1. Explain GET and POST methods.

Answer--> In web development, HTTP (Hypertext Transfer Protocol) is used to transmit data between the client and the server. GET and POST are two HTTP methods used to send and retrieve data from a server.

GET Method:
The GET method is used to request data from the server. When a user submits a GET request, the data is sent through the URL. The data is visible in the URL, which means that it can be bookmarked, cached, and shared. Because the data is transmitted through the URL, there is a limit to how much data can be sent. 

POST Method:
The POST method is used to send data to the server for processing. When a user submits a POST request, the data is sent in the body of the HTTP request. This means that the data is not visible in the URL, which makes it more secure. There is no limit to the amount of data that can be sent using the POST method. 

Q2. Why is request used in Flask?

Answer--> In Flask, the request object is used to access incoming data from a client request. When a client sends a request to a Flask application, the request object is automatically created and populated with data from the client's request.

Here are some of the most common uses of request object in Flask:

1 Accessing query parameters: The request object allows you to access query parameters in the URL.

For example, if the client navigates to http://example.com/search?q=Flask, you can access the value of the q parameter using request.args['q']. 

2 Accessing form data: The request object allows you to access form data submitted by the client. 

For example, if you have a form with an input field named name, you can access the value of that field using request.form['name'].



3 Accessing cookies: The request object allows you to access cookies that have been set by the client.

For example, if the client has a cookie named username, you can access the value of that cookie using request.cookies['username'].

Q3. Why is redirect() used in Flask?

Answer--> In Flask, the redirect() function is used to redirect the user to a different endpoint or URL. It is used when you want to redirect the user to a different page after they have performed a certain action, such as submitting a form.

Here are some of the most common use cases for the redirect() function in Flask:

1 After a successful form submission: If a user submits a form successfully, you can redirect them to a "thank you" page or to the page that displays the submitted data.

2 After a failed form submission: If a user submits a form with errors, you can redirect them back to the form page with an error message.

3 After a login/logout: If a user logs in or logs out of your application, you can redirect them to the appropriate page.

4 After a resource creation or update: If a user creates or updates a resource in your application, you can redirect them to the page that displays the updated resource.
